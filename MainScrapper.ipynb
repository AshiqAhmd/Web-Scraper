```python
def extract_url(url):
    
    driver.get(url)
    time.sleep(5)
    
    content = driver.page_source
    soup = BeautifulSoup(content)
    
    li=[]
    a=soup.findAll('a',href=True,attrs={'class':'product-item-list__images'})
    try:
        a=a[0].get("href")
        li.append("https://boodmo.com"+a)
    except:
        print("no url")
    return li
```
